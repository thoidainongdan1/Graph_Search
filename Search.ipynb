# DFS Search

```python
from collections import deque

def dfs(graph, start, goal):
    visited = set() # Các nút đã được duyệt
    trace = {} # Lưu vết đường đi
    stack = deque() # Tập biên O
    temp = deque() # Tác dụng hiển thị 
    stack.append(start)
    temp.append(start)
    trace[start] = 0
    print("STT\tNút được mở rộng\tTập biên O")
    id = 0
    print(id, list(stack), sep = '\t\t\t\t')

    while stack: # Lặp tới khi stack rỗng
        id += 1
        node = stack.popleft() # Lấy đỉnh stack rồi loại bỏ đỉnh đó ra khỏi stack
        node_temp = temp.popleft() # Tác dụng hiển thị
        
        if node not in visited: # Kiểm tra đỉnh đang xét đã được duyệt chưa
            visited.add(node)

            if node == goal: # Đỉnh đang xét là đích
                print(id, node_temp, "Đích", sep='\t\t')   
                break
            
            for neighbor in graph[node][::-1]: # Duyệt các đỉnh kề với đỉnh đang xét
                if neighbor not in visited: # Kiểm tra các đỉnh đó đã được duyệt chưa
                    stack.appendleft(neighbor) # Đẩy neighbor vào đỉnh stack
                    temp.appendleft(neighbor+"("+node+")") # Tác dụng hiển thị
                    trace[neighbor] = node # Lưu vết
                    
        print(id, node_temp, list(temp), sep='\t\t') # In trạng thái   
    
    # In đường đi
    print("\n============================================================================\n")
    print("Đường đi: ", end = "")
    print(goal, end = "")
    while trace[goal] != 0:
        print(" <-", trace[goal], end = "")
        goal = trace[goal]
```

# BFS Search

```python
from queue import Queue

def bfs(graph, start, goal):
    visited = set() # Các nút đã được duyệt
    trace = {} # Lưu vết đường đi
    q = Queue() # Tập biên O
    temp = Queue() # Tác dụng hiển thị
    q.put(start)
    temp.put(start)
    trace[start] = 0
    print("STT\tNút được mở rộng\tTập biên O")
    id = 0
    print(id, list(q.queue), sep = '\t\t\t\t')

    while not q.empty():
        id += 1
        node = q.get() # Lấy và loại bỏ đỉnh ở đầu hàng đợi
        node_temp = temp.get() # Tác dụng hiển thị
        
        if node not in visited: # Kiểm tra đỉnh đã được duyệt chưa
            visited.add(node)

            if node == goal: # Đỉnh đang xét là đích
                print(id, node_temp, "Đích", sep='\t\t')   
                break
            
            for neighbor in graph[node]: # Duyệt các đỉnh kề với đỉnh đang xét
                # Kiểm tra các đỉnh đó đã được duyệt chưa và có thuộc tập O không
                if neighbor not in visited and neighbor not in list(q.queue): 
                    q.put(neighbor) # Đưa neighbor vào đầu hàng đợi
                    temp.put(neighbor+"("+node+")") # Tác dụng hiển thị
                    trace[neighbor] = node # Lưu vết
                    
        print(id, node_temp, list(temp.queue), sep='\t\t')  
            
    # In đường đi
    print("\n============================================================================\n")
    print("Đường đi: ", end = "")
    print(goal, end = "")
    while trace[goal] != 0:
        print(" <-", trace[goal], end = "")
        goal = trace[goal]        
```

# IDS Search

```python
from collections import deque

def ids(graph, start, goal):
    trace = {} # Lưu vết đường đi
    trace[start] = 0
    def dfs(graph, start, goal, depth, trace):
        visited = set() # Các nút đã được duyệt
        stack = deque() # Tập biên O
        temp = deque() # Tác dụng hiển thị 
        stack.append((0, start)) # Stack chứa độ sâu của đỉnh và tên đỉnh
        temp.append(start)
        id = 0
        print("i =", depth)
        print("STT\tNút được mở rộng\tTập biên O")
        print(id, list(stack), sep = '\t\t\t\t')  
        
        while stack:            
            id += 1
            cur_depth, node = stack.popleft() # Lấy đỉnh stack rồi loại bỏ đỉnh đó ra khỏi stack
            node_temp = temp.popleft() # Tác dụng hiển thị

            if node not in visited: # Kiểm tra đỉnh đã được duyệt chưa
                visited.add(node)

                if node == goal: # Đỉnh đang xét là đích
                    print(id, node_temp, "Đích", sep='\t\t')   
                    return True # True khi tìm thấy đích

                if cur_depth < depth: # Chưa đạt độ sâu tối đa
                    for neighbor in graph[node][::-1]: # Duyệt các đỉnh kề với đỉnh đang xét
                        if neighbor not in visited: # Kiểm tra đỉnh đó đã được duyệt chưa
                            stack.appendleft(((cur_depth+1), neighbor)) # Đẩy độ sâu và neighbor vào đỉnh stack
                            temp.appendleft(neighbor+"("+node+")") # Tác dụng hiển thị
                            trace[neighbor] = node # Lưu vết

            print(id, node_temp, list(temp), sep='\t\t') # In trạng thái
        return False # False khi chưa tìm thấy đích      
    
    # IDS search + hiển thị
    depth = 0
    flag = dfs(graph, start, goal, depth, trace)
    print("\n============================================================================\n")
    while(flag is False):
        depth += 1  
        flag = dfs(graph, start, goal, depth, trace)
        print("\n============================================================================\n")
        
    # In đường đi
    print("Đường đi: ", end = "")
    print(goal, end = "")
    while trace[goal] != 0:
        print(" <-", trace[goal], end = "")
        goal = trace[goal]
```

# UCS Search

```python
class Graph:
    '''
    Graph gồm:
        - nodes: Các đỉnh trong đồ thị
        - edges: Danh sách kề của các đỉnh
        - weights: Chi phí ứng với các cạnh
    '''
    
    def __init__(self, edges, weights): # Khởi tạo
        self.edges = edges
        self.weights = weights

    def neighbors(self, node): # Đưa ra danh sách cách đỉnh kề với node
        return self.edges[node]

    def get_cost(self, from_node, to_node): # Đưa ra chi phí khi đi từ from_node đến to_node
        return self.weights[(from_node, to_node)]
    
def ucs(graph, start, goal):
    visited = set() # Các nút đã được duyệt
    queue = [(0, start)] # queue chứa chi phí khi đi từ start tới 1 đỉnh và tên đỉnh đó
    trace = {} # Lưu vết đường đi
    trace[start] = 0
    final_cost = -1 # Lưu chi phí tối ưu
    print("STT\tNút được mở rộng\tTập biên O")
    id = 0
    print(id, "(0, "+start+")", sep = '\t\t\t\t')

    while queue:
        id += 1
        cost, node = min(queue) # Chọn ra đỉnh có chi phí nhỏ nhất trong queue 
        queue.remove((cost, node)) # Xoá đỉnh đó khỏi queue
        
        if node not in visited: # Kiểm tra đỉnh đã được duyệt chưa
            visited.add(node)

            if node == goal: # Đỉnh đang xét là đích
                final_cost = cost # Cập nhật chi phí tối ưu
                print(id, node+"("+str(trace[node])+")", "Đích", sep='\t\t')   
                break
            
            for i in graph.neighbors(node): # Duyệt các đỉnh kề với đỉnh đang xét
                if i not in visited: # Kiểm tra đỉnh đó đã được duyệt chưa
                    total_cost = cost + graph.get_cost(node, i) # Cập nhật chi phí tới đỉnh i
                    flag = 0 # neighbor không có trong queue
                    for c, n in queue:
                        if i == n: # Tìm được đỉnh i trong queue
                            if total_cost < c:
                                flag = 1 # neighbor có trong queue và chi phí mới tốt hơn
                                queue.remove((c, n)) # Xoá neighbor trong queue
                            else:
                                flag = 2 # neighbor có trong queue và chi phí mới kém hơn
                            break
                    if flag == 0 or flag == 1:
                        queue.append((total_cost, i)) # Đưa chi phí và đỉnh vào queue
                        trace[i] = node # Lưu vết
        
        # In trạng thái
        print(id, node+"("+str(trace[node])+")", sep = '\t\t', end = "") 
        print("\t\t", end = "")
        for i in queue:
            print(str(i)+"("+trace[i[1]]+")", end = "   ")
        print()    
        
    # In đường đi
    print("\n============================================================================\n")
    print("Đường đi: ", end = "")
    print(goal, end = "")
    while trace[goal] != 0:
        print(" <-", trace[goal], end = "")
        goal = trace[goal] 
    print("\nCost =", final_cost)   
```
